```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
customer_df = pd.read_csv("https://raw.githubusercontent.com/dicodingacademy/dicoding_dataset/main/DicodingCollection/customers.csv")
customer_df.name = "customer"
customer_df.head()
```

```python
orders_df = pd.read_csv("https://raw.githubusercontent.com/dicodingacademy/dicoding_dataset/main/DicodingCollection/orders.csv")
orders_df.name = "orders"
orders_df.head()
```

```python
product_df = pd.read_csv("https://raw.githubusercontent.com/dicodingacademy/dicoding_dataset/main/DicodingCollection/products.csv")
product_df.name = "product"
product_df.head()
```

```python
sales_df = pd.read_csv("https://raw.githubusercontent.com/dicodingacademy/dicoding_dataset/main/DicodingCollection/sales.csv")
sales_df.name = "sales"
sales_df.head()
```

#Data Assessment

```python
# prompt: create array cointaining this variable and don't generate the import function:
# customer_df
# orders_df
# product_df
# sales_df

dataframes = [customer_df, orders_df, product_df, sales_df]

for df in dataframes:

   print(f"---data {df.name} info---")
   print(df.info())

   print("---data missing---")
   print( df.isna().sum())

   print("---data duplicated---")
   print(df.duplicated().sum())

   print("---data Desc---")
   print(df.describe())

```

#Data Cleaning

```python
#Duplicated Data
for df in dataframes:
   df.drop_duplicates(inplace=True)

for df in dataframes:
   print(f"---data {df.name} info---")
   print(df.duplicated())
```

```python
#Missing Data

for df in dataframes:
   print(f"---data {df.name} info---")
   print(df.isna().sum())
```

```python
#Customer Missing values
print(customer_df.gender.value_counts())
customer_df.gender.fillna(value="Prefer not to say", inplace=True)
print(customer_df.gender.isna())
```

```python
#Salse missing value

print(sales_df.total_price.isna().sum())
sales_df.total_price.fillna(value=sales_df.price_per_unit * sales_df.quantity, inplace=True)
```

```python
print(sales_df.total_price.isna().sum())
```

```python
#Inacurate Data
for ages in customer_df.age:
  if ages == customer_df.age.max():
    print(ages)
```

```python
customer_df.age.replace(customer_df.age.max(), 70, inplace=True)
```

```python
customer_df.age.replace({customer_df.age.max(): 50}, inplace=True)
```

#Data Exploratory

```python
customer_df.groupby(by="city").customer_id.nunique().sort_values(ascending=False)
```

```python
customer_df.groupby(by="state").customer_id.nunique().sort_values(ascending=False)
```

```python
orders_df["order_date"] = pd.to_datetime(orders_df.order_date)
orders_df["delivery_date"] = pd.to_datetime(orders_df.delivery_date)
delivery_time = orders_df["delivery_date"] - orders_df["order_date"]
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
orders_df["delivery_time"] = round(delivery_time/86400)
```

```python
orders_df["delivery_time"]
```

```python
orders_df.describe(include="all")
```

```python
customer_id_in_orders_df =  orders_df.customer_id.tolist()
customer_df["status"] = customer_df["customer_id"].apply(lambda x: "Active" if x in customer_id_in_orders_df else "Non Active")
customer_df.sample(5)
```

```python
#Active|Non active in percentage
(customer_df.groupby(by="status").customer_id.count() / customer_df.customer_id.count()) * 100
```

```python
orders_customers_df = pd.merge(
    left=orders_df,
    right=customer_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
orders_customers_df.head()
```

```python
orders_customers_df.groupby(by="city").order_id.nunique().sort_values(ascending=False).reset_index().head(10)
```

```python
(orders_customers_df.groupby(by="gender").order_id.nunique().sort_values(ascending=False) / orders_customers_df.order_id.count()) * 100
```
